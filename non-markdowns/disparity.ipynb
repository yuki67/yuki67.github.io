# 格差のできる様子を観察する

いつだか受けた統計力学の講義で見た「平等な取引を繰り返すだけで格差ができる」ことのデモが印象に残っていたので自分で作りました。

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np
plt.rcParams['figure.figsize'] = (10, 6) 
```

```python
n = 50
persons = range(n)
moneys = np.ones(n) * 100
plt.bar(persons, moneys)
plt.show()
print(moneys.std())
```

```python
import random

n = 100
persons = range(n)
moneys = np.ones(n) * 100

for i in range(100000):
    loser = random.randint(0, n-1)
    winner = random.randint(0, n-1)
    moneys[loser] -= 1
    moneys[winner] += 1

plt.bar(persons, sorted(moneys))
plt.show()
print(moneys.std())
```

```python
import random

n = 100
persons = range(n)
moneys = np.ones(n) * 100

for i in range(10000):
    winner = random.randint(0, n-1)
    pay = random.randint(0, n-1)
    moneys[winner] += pay
    for k in range(pay):
        moneys[random.randint(0, n-1)] -= 1
            

plt.bar(persons, sorted(moneys))
plt.show()
print(moneys.std())
```

```python
import random

n = 100
persons = range(n)
moneys = np.ones(n) * 100

for i in range(100000):
    loser = random.randint(0, n-1)
    winner = random.randint(0, n-1)
    moneys[loser] *= 0.99
    moneys[winner] *= 1.01

plt.bar(persons, sorted(moneys))
plt.show()
print(moneys.std())
```

```python
import random

n = 50
persons = range(n)
moneys = np.ones(n) * 100

for i in range(100000):
    losers = random.randint(0, n-1), random.randint(0, n-1)
    winner = random.randint(0, n-1)
    moneys[losers[0]] -= 1
    moneys[losers[1]] -= 1
    moneys[winner] += 2

plt.bar(persons, sorted(moneys))
plt.show()
```
